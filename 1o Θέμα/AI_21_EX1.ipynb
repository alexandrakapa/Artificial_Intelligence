## Κώδικας visualization


````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[~grid.grid] = [0, 0, 0]
        image[grid.grid] = [255, 255, 255]
        # Use this to treat 1/True as obstacles
        # image[grid.grid] = [0, 0, 0]
        # image[~grid.grid] = [255, 255, 255]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
%matplotlib inline
import numpy as np
import queue
import  random 
import matplotlib.pyplot as plt
import copy


class Maze:
    def __init__(self, N, S, F):

      """
      N: integer that indicates the size of the NxN grid of the maze
      S: pair of integers that indicates the coordinates of the starting point (S)
      F: pair of integers that indicates the coordinates of the finish point (F)
      You can add any other parameters you want to customize maze creation (e.g. variables that
      control the creation of additional paths)
      """

      assert N > 2

      ## Make sure start and end are within the grid

      assert S[0] < N-1
      assert S[1] < N-1
      assert F[0] < N-1
      assert F[1] < N-1

      assert S[0] > 0
      assert S[1] > 0
      assert F[0] > 0
      assert F[1] > 0

      # Add here any additional constraints your implementation may have
      assert N % 2 == 1
      # Require S, F to be in "cell" positions 
      assert S[0] % 2 == 1
      assert S[1] % 2 == 1
      assert F[0] % 2 == 1
      assert F[1] % 2 == 1

      self.N = N
      self.S = S
      self.F = F

      # Grid initialized with obstacles (array of 0/False)
      # 1/True indicates available cells
      self.grid = np.zeros((N, N), dtype=bool)
      
      ## YOUR CODE HERE
      visited = copy.deepcopy(self.grid)

      self.grid[F[0], F[1]] = True
      # Pick starting cell
      cell = self.S
      self.grid[cell[0], cell[1]] = True
      visited[cell[0], cell[1]] = True
      # Add its walls to a list (list of pairs -> list of lists)
      wallList = [ [cell[0]+1, cell[1]] , [cell[0]-1, cell[1]] , [cell[0], cell[1]+1] , [cell[0], cell[1]-1] ]

      while(len(wallList) > 0):
        # pick random wall and remove from list
        random_index = random.randrange(len(wallList))
        wall = wallList.pop(random_index)
        # a wall (x,y) where x is odd is a horizontal wall connecting (x, y-1) - (x, y+1)
        # a wall (x,y) where x is even is a vertical wall connecting (x-1, y) - (x+1, y)
        
        # if picked wall is on the borders
        if(wall[0] <= 0 or wall[0] >= self.N-1 or wall[1] <= 0 or wall[1] >= self.N-1):
          continue

        # In case of vertical wall
        if(wall[0] % 2 == 0):
          # if both are visited then move on
          if(visited[wall[0]-1 , wall[1]] and visited[wall[0]+1 , wall[1]]):
            continue

          # break the wall
          self.grid[wall[0], wall[1]] = True

          # mark the unvisited cell as visited
          # if one is visited then the other is not
          if(visited[wall[0]-1 , wall[1]]):
            cell = [wall[0]+1 , wall[1]]
          else:
            cell = [wall[0]-1 , wall[1]]
          
          self.grid[cell[0], cell[1]] = True
          visited[cell[0], cell[1]] = True

          # Add the neighboring walls of the cell (that are not broken) to the wall list
          candWalls = [ [cell[0]+1, cell[1]] , [cell[0]-1, cell[1]] , [cell[0], cell[1]+1] , [cell[0], cell[1]-1] ]

          for i in candWalls:
            # if the wall is not broken
            if(self.grid[i[0], i[1]] == False):
              wallList.append(i)
        
        # In case of horizontal wall
        else:
          # if both are visited then move on
          if(visited[wall[0] , wall[1]+1] and visited[wall[0] , wall[1]-1]):
            continue

          # break the wall
          self.grid[wall[0], wall[1]] = True

          # mark the unvisited cell as visited
          cell = [wall[0], wall[1]] # just an initaialization
          
          # if one is visited then the other is not
          if(visited[wall[0] , wall[1]-1]):
            cell = [wall[0] , wall[1]+1]
          else:
            cell = [wall[0] , wall[1]-1]

          self.grid[cell[0], cell[1]] = True
          visited[cell[0], cell[1]] = True

          # Add the neighboring walls of the cell (that are not broken) to the wall list
          candWalls = [ [cell[0]+1, cell[1]] , [cell[0]-1, cell[1]] , [cell[0], cell[1]+1] , [cell[0], cell[1]-1] ]

          for i in candWalls:
            # if the wall is not broken
            if(self.grid[i[0], i[1]] == False):
              wallList.append(i)
      
      # the following is about making more paths to final cell
      bfs1 = self.BFS(self.grid, self.S)
      bfs2 = self.BFS(self.grid, self.F)

      (wallH, wallV) = self.WallToBreak(self.grid, bfs1, bfs2)
      # print(wallH, wallV)
      assert wallH or wallV
      if wallH:
        self.grid[wallH] = True
      if wallV:
        self.grid[wallV] = True
        
    def BFS(self, grid, start):
        # distance[x,y] = 0 -> there is a wall at [x,y] 
        distances = copy.deepcopy(grid)
        distances = distances.astype('int')
        # mark all wall cells as visited
        visited = np.invert(grid)

        visited[start] = True
        distances[start] = 1
        Q = queue.Queue()
        Q.put(start)
        while not Q.empty():
          cell = Q.get()
          for neighbor in (cell[0] + 1, cell[1]), (cell[0] - 1, cell[1]), (cell[0], cell[1] + 1), (cell[0], cell[1] - 1):
            if visited[neighbor]:
              continue

            visited[neighbor] = True
            distances[neighbor] = distances[cell] + 1
            Q.put(neighbor) 

        return distances

    def WallToBreak(self, grid, bfs1, bfs2):
        coloring = bfs1 - bfs2

        # we look for reginons in coloring talbe where [a, b, c] (vertically or horizontally): 
        # b = 0 (wall or both bfs get there simultaneously)
        # a * c < 0 (one bfs gets there faster than the other)

        # find a horizontal breakable wall
        horizontalWall  = None
        step = 0
        mid = self.N // 2 
        row = mid
        depth = 1
        while not horizontalWall and row > 0 and row < self.N - 1:
          # print('Checking row ', row)
          for col in range(1, self.N - 3):
            a = coloring[row, col]
            wallCheck = (row, col+1)
            c = coloring[row, col+2]

            # print('Considering wall ', wallCheck)
            if(grid[wallCheck] == False and a * c < 0):
              # we found a breakable wall
              horizontalWall = wallCheck
              break

          if step % 2 == 0:
            row = mid - depth
          else:
            row = mid + depth
            depth += 1
          step += 1


        # find a vertical breakable wall
        verticalWall  = None
        step = 0
        mid = self.N // 2 
        col = mid
        depth = 1
        while not verticalWall and col > 0 and col < self.N - 1:
          # print('Checking col ', col)
          for row in range(1, self.N - 3):
            a = coloring[row, col]
            wallCheck = (row+1, col)
            c = coloring[row+2, col]

            # print('Considering wall ', wallCheck)
            if(grid[wallCheck] == False and a * c < 0):
              # we found a breakable wall
              verticalWall = wallCheck
              break

          if step % 2 == 0:
            col = mid - depth
          else:
            col = mid + depth
            depth += 1
          step += 1
        
        return (horizontalWall, verticalWall)


    def draw_map(self, path=None):
        """ Draws the maze as an image. Considers grid values of 0/False to represent obstacles and
        values of 1/True to represent empty cells, but this can be customized. Obstacles are painted
        black and empty cells are painted white. Starting point is painted green and finish point red.
        Optionally accepts as a parameter a path within the maze which is painted blue. 
        """
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[~self.grid] = [0, 0, 0]
        image[self.grid] = [255, 255, 255]

        # Uncomment the next 2 lines of code to treat 1/True as obstacles (and 0/False as free maze cells)
        # image[self.grid] = [0, 0, 0]
        # image[~self.grid] = [255, 255, 255]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
````

*Επίλυση λαβυρίνθου:*

````
class Pathfinder:
    def __init__(self, maze, c, h):
        """
        maze : Αντικείμενο τύπου Maze (από Μέρος 1)
        c : Συνάρτηση που υπολογίζει την πραγματική απόσταση μεταξύ δύο σημείων
        h : Συνάρτηση που υπολογίζει την ευριστική μεταξύ δύο σημείων
        """
        self.maze = maze
        self.vis = visualization(maze.S, maze.F)
        self.path = []
        self.cost = c
        self.heuristic = h

        self.expanded_nodes = []
        
        ### Fill the path list with the coordinates of each point in the path from maze.S to maze.F
        ### Your code here
        open = []       # The set of nodes to be evaluated
        closed = []     # The set of nodes already evaluated
        # g = maze.BFS(maze.grid, maze.S)
        g = np.zeros((maze.N, maze.N), dtype='int')
        g[maze.S] = 1
        open.append(( 0, maze.S, None))

        # Items in open and closed lists are tuples: (f value, point, parent) -> (38, (5,3), (4,3))
        while len(open) > 0:
          # remove item with minimum f value from open list
          # and place it in closed list
          current = min(open)
          open.remove(current)
          closed.append(current)

          # open = [(38, (5,3), (4,3)), (21, (1,2), (2,1))]
          open2 = [item[1] for item in open]
          closed2 = [item[1] for item in closed]
          self.vis.draw_step(maze, open2, closed2)
          
          self.expanded_nodes = list(closed2)

          if current[1] == maze.F:
            # path has been found
            break
          
          for neighbor in self.neighbors(current[1]):
            parent = current[1]
            # check if neighbor exists in closed set
            exists = next((i for i, v in enumerate(closed) if v[1] == neighbor), None)
            # if neighbor is not traversable or neighbor is in closed set
            if self.maze.grid[neighbor] == False or exists:
              continue
            
            # compute f_value 
            # f_val = h
            f_val = self.heuristic(neighbor, self.maze.F)
            # if c(neighbor[0], neighbor[1]) > 0:
            #   f_val += g[neighbor]
            new_g = g[parent] + c(neighbor, parent)
            if new_g < g[neighbor] or g[neighbor] == 0:
              g[neighbor] = new_g
            f_val += g[neighbor]
            # check if neighbor exists in open set
            index_in_open = next((i for i, v in enumerate(open) if v[1] == neighbor), None)
            if index_in_open:
              # if new path to neighbor is shorter 
              # update f value and parent cell
              if f_val < open[index_in_open][0]:
                open[index_in_open][0] = f_val
                open[index_in_open][2] = parent
            else:
              new_cell = ( f_val, neighbor, parent)
              open.append(new_cell)

        # print(g)
        # backtrack from target to create path
        current = self.maze.F
        while True:
          self.path.insert(0, current)
          if current == self.maze.S:
            break
          index_in_closed = next((i for i, v in enumerate(closed) if v[1] == current), None)
          parent = closed[index_in_closed][2]
          current = parent
        
        self.vis.add_path(self.path)
    
    def neighbors(self, cell):
      return [(cell[0] + 1, cell[1]) , (cell[0] - 1, cell[1]) , (cell[0], cell[1] + 1) , (cell[0], cell[1] - 1)]

    def get_path(self):
        return self.path
    
    # A function to compute path length
    def path_length(self):
        return len(self.path)
    
    # A function to return the number of expanded nodes
    def expanded_nodes_length(self):
      return len(self.expanded_nodes)
````

*Ευριστικές συναρτήσεις h(n):*

````
def manhattan_distance(start, target):
    return abs(start[0] - target[0]) + abs(start[1] - target[1])

def euclidean_distance(start, target):
    return np.sqrt((start[0] - target[0])**2 + (start[1] - target[1])**2)

def zero_distance(start, target):
    return 0
    
#Υλοποιήσαμε επίσης και τις chepyshev και octile distances:

def chebyshev_distance(start, target):   
    return abs(start[0] - target[0]) + abs(start[1] - target[1]) - min(abs(start[0] - target[0]), abs(start[1] - target[1]))

def octile_distance(start, target):
    return abs(start[0] - target[0]) + abs(start[1] - target[1]) + (np.sqrt(2)-2) * min(abs(start[0] - target[0]), abs(start[1] - target[1]))

print("Manhattan distance: ",manhattan_distance((1,3),(5,7)))
print("Euclidean distance: ",euclidean_distance((1,3),(5,7)))
print("Chebyshev distance: ",chebyshev_distance((1,3),(5,7)))
print("Octile distance: ",octile_distance((1,3),(5,7)))
````

#### Κατασκευή γραφικής παράστασης ***μήκους ελαχίστου μονοπατιού συναρτήσει του μεγέθους του χάρτη***

Για την δημιουργία της γραφική παράστασης υπολογίσαμε για κάθε αλγόριθμο τον μέσο όρο του μήκους του ελάχιστου μονοπατιού 10 λαβυρίνθων ίδιου μεγέθους για κάθε $$Ν ∈ [11,51]$$
Όταν στον αλγόριθμο έχουμε διαφορετικούς συνδυασμούς αποστάσεων, βρήκαμε τον μέσο όρο των αποτελεσμάτων όλων των ειδών των αποστάσεων (συμπεριλαμβάνοντας και τις επιπλέον αποστάσεις που ορίσαμε εμείς δηλαδή chebyshev και octile).

````
best_first_manhattan = np.zeros(10)
best_first_euclidean = np.zeros(10)
best_first_chebyshev = np.zeros(10)
best_first_octile = np.zeros(10)
dijkstra = np.zeros(10)
a_manhattan = np.zeros(10)
a_euclidean = np.zeros(10)
a_chebyshev = np.zeros(10)
a_octile = np.zeros(10)

best_first_average_length = np.zeros(20)
dijkstra_average_length = np.zeros(20)
a_average_length = np.zeros(20)

best_first_path = 0
dijkstra_path = 0
a_path = 0

i = 0
j = 0
# Map size : all the odd numbers from 11 to 51 
for N in range (11,51,2):
      S = (1, 1)
      F = (N-2, N-2)
      #We want 10 different maps for every map size
      while (i < 10):
      # We use exceptions so that if a random maze is not valid we ignore it and make a new one
       try :
           # Create a new maze (the algorithms should run on the same labyrinths)
           map = Maze(N, S, F)

           # For Best-First algorithm
           # Manhattan distance
           pf_m = Pathfinder(map, c = lambda x, y: 0, h = manhattan_distance)
           best_first_manhattan[i] = pf_m.path_length()
           best_first_path += best_first_manhattan[i]
           # Euclidean distance
           pf_e = Pathfinder(map, c = lambda x, y: 0, h = euclidean_distance)
           best_first_euclidean[i] = pf_e.path_length()
           best_first_path += best_first_euclidean[i]
           # Chebyshev distance
           pf_c = Pathfinder(map, c = lambda x, y: 0, h = chebyshev_distance)
           best_first_chebyshev[i] = pf_c.path_length()
           best_first_path += best_first_chebyshev[i]
           # Octile distance
           pf_o = Pathfinder(map, c = lambda x, y: 0, h = octile_distance)
           best_first_octile[i] = pf_o.path_length()
           best_first_path += best_first_octile[i]

           # For Dijkstra algorithm
           pf_d = Pathfinder(map, c = lambda x, y: 1, h = zero_distance)
           dijkstra[i] = pf_d.path_length()
           dijkstra_path += dijkstra[i]

           # For A* algorithm
           # Manhattan distance
           pf_a_m = Pathfinder(map, c = lambda x, y: 1, h = manhattan_distance)
           a_manhattan[i] = pf_a_m.path_length()
           a_path += a_manhattan[i]
           # Euclidean distance
           pf_a_e = Pathfinder(map, c = lambda x, y: 1, h = euclidean_distance)
           a_euclidean[i] = pf_a_e.path_length()
           a_path += a_euclidean[i]
           # Chebyshev distance
           pf_a_c = Pathfinder(map, c = lambda x, y: 1, h = chebyshev_distance)
           a_chebyshev[i] = pf_a_c.path_length()
           a_path += a_chebyshev[i]
           # Octile distance
           pf_a_o = Pathfinder(map, c = lambda x, y: 1, h = octile_distance)
           a_octile[i] = pf_a_o.path_length()
           a_path += a_octile[i]

           i += 1
       except AssertionError:
          #  print("Error! ")
          continue
      # print(best_first_manhattan)
      # print(best_first_euclidean)

      # Compute the average for these algorithms
      best_first_average_length[j] = np.int_(best_first_path)/40
      dijkstra_average_length[j] = np.int_(dijkstra_path)/10
      a_average_length[j] = np.int_(a_path)/40
      print("Best first average length:",best_first_average_length[j],", Dijkstra average length:", dijkstra_average_length[j], ", A* average length ", a_average_length[j])
      # print(dijkstra_average_length[j])
      j += 1
      # For the next iteration
      i = 0
      best_first_manhattan = np.zeros(10)
      best_first_euclidean = np.zeros(10) 
      best_first_chebyshev = np.zeros(10)
      best_first_octile = np.zeros(10)
      dijkstra = np.zeros(10)
      a_manhattan = np.zeros(10)
      a_euclidean = np.zeros(10) 
      best_first_path = 0
      dijkstra_path = 0
      a_path = 0
````

Με βάση αυτό το σύστημα συντεταγμένων ορίζουμε ως το ζεύγος (x,y) τις συντεταγμένες κάθε θέσης του λαβυρίνθου με x να αφορά την οριζόντια συντεταγμένη και y την κάθετη. Για την εκτέλεση του hill-climbing θα χρησιμοποιήσουμε την απόσταση Manhattan η οποία ορίζεται ως $$d((x1,y1),(x2,y2)) = |x1-x2| + |y1-y2|$$\
Ο αλγόριθμος hill-climbing σε κάθε βήμα διαλέγει το γειτονικό του κελί με την μικρότερη απόσταση Manhattan από τον στόχο και ορίζει το νέο αυτό κελί ως current node. Σε περίπτωση ισοβαθμίας των αποστάσεων διαλέγεται τυχαία ένα από τα κελιά που έχουν ίδια απόσταση από τον στόχο και ο αλγόριθμος συνεχίζεται.

Πιο αναλυτικά, χρησιμοποιώντας όλα τα παραπάνω μπορεί να οριστεί το δέντρο εκτέλεσης του αλγορίθμου hill-climbing όπως παρακάτω:

Ως εκθέτης σε κάθε ζεύγος ορίζεται η Manhattan απόσταση του από το κελί-στόχος και ως ροζ ορίζονται τα κελιά που επιλέγει ο αλγόριθμος hill-climbing.

Παρατηρούμε ότι ο αλγόριθμος σταματάει όταν φτάσει στο κελί (3,5) μιας και δεν μπορεί να συνεχίσει αφού το μοναδικό κελί στο οποίο μπορεί να συνεχίσει είναι το (3,6) το οποίο έχει μεγαλύτερη απόσταση Manhattan από ότι το current κελί. Στην αρχή της εκτέλεσης του αλγορίθμο μπορούμε να επιλέξουμε οποιοδήποτε από τα (8,9) και (9,8) μιας και έχουν την ίδια απόσταση,ωστόσο διαλέγουμε τυχαία το (9,8).\

#### Σχολιασμός

Προκειμένου να αργεί όσο το δυνατόν περισσότερο ο Α*, θα πρέπει να κάνει συνεχώς backtracking, δηλαδή είτε να διαλέγει διαδρομές που οδηγούν σε αδιέξοδο είτε να καταλήγει σε κελιά που έχουν μεγαλύτερη f value από προηγούμενα κελιά που έχει διακόψει την εξερεύνηση τους. Υπενθυμίζουμε ότι ο Α * βασίζεται στην συνάρτηση κόστους $$f(n) = g(n) + h(n))$$
Επομένως θα πρέπει ο λαβύρινθος μας να καταλήγει σε πολλά αδιέξοδα προκειμένου να αργεί όσο το δυνατόν περισσότερο ο Α *.

Αρχικά αριθμούμε τον λαβύρινθο από αρίστερα προς τα δεξιά και από πάνω προς τα κάτω(αγνοώντας κατά την αρίθμηση το μαύρο πλαίσιο):

![](https://drive.google.com/uc?export=view&id=109EixEt5NSg2obCLvONJNicYUPVeB5Hx)

Με βάση την δομή του παραπάνω λαβυρίνθου, βλέπουμε ότι ενώ για όλα τα μονοπάτια (πλην του μοναδικού που φτάνει στον στόχο) φαίνεται ότι ακολουθώντας τα πλησιάζουμε στον τελικό στόχο, τελικά καταλήγουμε σε αδιέξοδο και επομένως καθυστερούμε. Επίσης παρατηρούμε ότι το μοναδικό μονοπάτι που μας οδηγεί στον στόχο είναι αυτό που έχει το μεγαλύτερο μήκος και επομένως εύλογα αντιλαμβανόμαστε ότι αργούμε περισσοτερο να το διασχίσουμε. 

Πιο αναλυτικός σχολιασμός:

Κατα τα γνωστά με βάση τα παραπάνω προκύπτει το ακόλουθο δέντρο εκτέλεσης του αλγορίθμου.\
Κάθε κόμβος έχει την μορφή $$(x,y) (g(n) + h(n) = f(n), stepx)$$ όπου:
* (x,y) : είναι το ζεύγος συντεταγμένων
* g(n),h(n) : η πραγματική και η ευριστική απόσταση αντίστοιχα 
* stepx : η σειρά των βημάτων κατά την εκτέλεση του αλγορίθμου\
Όπως γνωρίζουμε, ο Α * κάθε φορά διαλέγει τον κόμβο με το μικρότερο f και σε περίπτωση που φτάσει σε αδιέξοδο ή φτάσει σε κόμβο με f μεγαλύτερο από προηγούμενα κελιά που έχει διακόψει την εξερεύνηση τους, κάνει κάποιου είδους backtracking (γνωρίζουμε ότι γενικα ο Α* δεν υποστηρίζει backtracking ωστόσο χρησιμοποιούμε αυτόν τον όρο καθώς είναι πιο κοντά στο τι γίνεται κατά την εκτέλεση του αλγορίθμου) το οποίο γίνεται αντιληπτό αν παρακολουθήσουμε την ροή των steps.


![](https://drive.google.com/uc?export=view&id=1HpwacMDuEcP-m3tueN0ZCT0uRqr5jKwU)

To μονοπάτι που βρίσκει είναι το εξής (χρησιμοποιήσαμε την απόσταση Manhattan αν και τα αποτελέσματα είναι ίδια και για την euclidean):

````
%%time
pf = Pathfinder(maze=g, c = lambda x, y: 1, h = manhattan_distance)
g.draw_map(pf.get_path())
````

Παρατηρώντας το gif παρακάτω φαίνεται ξεκάθαρα ότι κάνει συνεχώς backtracking σύμφωνα με το παραπάνω tree diagram και για αυτό τον λόγο ο Α * αργεί όσο το δυνατόν περισσότερο μιας και ουσιαστικά εξερευνεί όλο τον λαβύρινθο.

````
pf.vis.show_gif(fps=2)
````

Όπως έχουμε προαναφέρει, μία ευριστική συνάρτηση πρέπει να διαθέτει τα δύο παρακάτω χαρακτηριστικά:
* Να είναι αποδεκτή, δηλαδή να μην υπερεκτιμά την απόσταση μέχρι τον στόχο αλλιώς ο αλγόριθμος μπορεί να επιστρέφει μη βέλτιστη λύση.
* Να επιστρέφει την μεγαλύτερη δυνατή απόσταση μέχρι τον στόχο γιατί σε αντίθετη περίπτωση, η υποτίμηση της απόστασης ενός κελιού από τον στόχο μπορεί να οδηγήσει στην επιλογή αυτού του κελιού έναντι άλλων που θα μπορούσαν να οδηγήσουν σε καλύτερη λύση.

Γνωρίζουμε ότι και οι δύο αποστάσεις είναι αποδεκτές επομένως σίγουρα θα βρει ο αλγόριθμος το βέλτιστο μονοπάτι και στις δύο περιπτώσεις. Με βάση τα παραπάνω, αντιλαμβανόμαστε ότι μεταξύ euclidean και manhattan αποστάσεις, η απόσταση manhattan βρίσκει πιο γρήγορα το μονοπάτι σε λαβυρίνθους όπως τους παραπάνω (δηλαδή που μπορούμε κάθε φορά να κινηθούμε αριστερα, δεξιά, πάνω, κάτω), αποτελεί δηλαδή καλύτερη μετρική μιας και η euclidean θα υποτιμά την απόσταση όλων των κελιών από τον στόχο λόγω του πυθαγόρειου θεωρήματος.\
Όπως βλέπουμε και παρακάτω δηλαδή, η euclidean απόσταση είναι 3 ενώ η manhattan 4 από την σχέση:
$$euclidean = sqrt(4+4) = 3$$
$$manhattan = 2 + 2 = 4$$\
![](https://drive.google.com/uc?export=view&id=1hPcNHqFxXs67WILpbjA6Z_K3cwcHqkvN)



Προς απόδειξη των ισχυρισμών μας, ορίζουμε έναν τυχαίο λαβύρινθο και χρονομετρούμε με ποιά από τις δύο αποστάσεις θα βρεθεί πιο γρήγορα το βέλτιστό μονοπάτι:

````
N = 51
S = (1, 1)
F = (39, 37)
maze = Maze(N, S, F)
maze.draw_map()
````

*Με την απόσταση Manhattan:*

````
%%time
pf = Pathfinder(maze=maze, c = lambda x, y: 1, h = manhattan_distance)
````

````
maze.draw_map(pf.get_path())
````

*Με την απόσταση Euclidean:*

````
%%time
pf = Pathfinder(maze=maze, c = lambda x, y: 1, h = euclidean_distance)
````

````
maze.draw_map(pf.get_path())
````

Από όπου επιβεβαιώνονται όσα αναφέραμε παραπάνω σχετικά με τον χρόνο εκτέλεσης του αλγορίθμου με χρήση και των δύο αποστάσεων, ότι δηλαδή με την Manhattan απόσταση βρίσκεται πιο γρήγορα το βέλτιστο μονοπάτι (σε 257 ms) σε σχέση με την Euclidean (σε 556ms).

#### Σχολιασμός 

Όπως γνωρίζουμε, ο αλγόριθμος Dijkstra δεν λαμβάνει υπόψιν κάποια ευριστική συνάρτηση, παρά μόνο την απόσταση από το αρχικό κελί την οποία κάθε φορά προσπαθεί να ελαχιστοποιήσει. Γι'αυτό τον λόγο, το πλήθος των expanded nodes του Dijkstra είναι πάντα μεγαλύτερο από του Α* καθώς εξερευνεί μεγαλύτερο αριθμό κελιών. Επίσης, γνωρίζουμε ότι τόσο o Dijkstra όσο και ο Α * (με την χρήση των αποστάσεων που έχουν ήδη αναφερθεί) βρίσκουν το βέλτιστο μονοπάτι.

Προκειμένου να καταφέρνει ο Α* να βρει το μονοπάτι προς τον στόχο πιο γρήγορα από τον Dijkstra, εκμεταλλευόμαστε την ιδιότητα του Dijkstra να βρίσκει το συντομότερο μονοπάτι από το αρχικό κελί προς όλα τα κελιά του λαβυρίνθου. Έτσι, ορίσαμε κατάλληλα τα εμπόδια ώστε ο Dijkstra να αργεί ψάχνοντας να βρει τα μονοπάτια προς όλους τους κόμβους ενώ ο Α* έψαχνε στοχευμένα το κελί-στόχο απευθείας.

Ο λόγος για τον οποίο αργεί ο Dijkstra είναι εξαιτίας του εμποδίου που έχουμε βάλει. Πιο συγκεκριμένα, αν υπάρχει κάποιο κάθετο εμπόδιο ανάμεσα στο αρχικό κελί και στο τελικό (το οποίο δηλαδή "κρύβει" ειτε το αρχικό είτε το τελικό κελί), τότε ο Dijkstra δυσκολεύεται περισσότερο να φτάσει στον στόχο μιας και εξερευνεί όλα τα υπόλοιπα μονοπάτια πριν περάσει το εμπόδιο.
